### 1. 取出知乎問題發問時間

```python
# 1. 取出知乎問題發問時間

import requests
import json
headers = {'user-agent': 'my-app/0.0.1'}
r = requests.get('https://www.zhihu.com/api/v4/questions/55493026/answers',headers=headers)
response = r.text
```

```python

datas = json.loads(response)

type(datas)
```

```python
datas.keys()
```

```python
datas['paging']
```

```python
datas['data']
```

```python
is_end=datas['paging']['is_end']
is_end
```

```python
time_stamp =[]
```

```python
url_get = datas['paging']['next']
url_get
```

```python
for data in datas['data']:
    time_stamp.append(data['created_time'])
```

```python
time_stamp
```

```python
while is_end==False:

    r = requests.get(url_get,headers=headers)
    response = r.text
    datas = json.loads(response)
    for data in datas['data']:
        time_stamp.append(data['created_time'])
    url_get = datas['paging']['next']
    is_end=datas['paging']['is_end']
    print(url_get)
    print(is_end)

```

```python
print('討論串共',len(time_stamp),'筆')
```

# 討論串發問問題的時間

```python
#
time_stamp.sort()
from datetime import datetime
for time in time_stamp:
    time = datetime.fromtimestamp(time)
    print(time)
```

### 2. 取出第一筆與最後一筆回答的時間

```python
# 2. 取出第一筆與最後一筆回答的時間

'''
Your Code
'''
first_time_stamp=datetime.fromtimestamp(time_stamp[0])
last_time_stamp=datetime.fromtimestamp(time_stamp[-1])

print('第一筆時間為',first_time_stamp)
print('最後一筆時間為',last_time_stamp)
```
