```python
fh = open("64_72hr_CH.xml", "r",encoding='UTF-8')
xml = fh.read()
fh.close()
```

```python
doc = dict(xmltodict.parse(xml))
```

```python
type(doc)
```

使用File I/O開啟的xml檔案為str型態，經過xmltodict的解析後成為dict型態

```python
districts = doc['cwbopendata']['dataset']['locations']['location']
```

```python
datasetDescription = doc['cwbopendata']['dataset']['datasetInfo']['datasetDescription']
```

```python
#顯示高雄市每一個地區所記錄的第一個時間點跟溫度
for district in districts:
    print (district["locationName"])
    print (district['weatherElement'][0]['description'])
    print (district['weatherElement'][0]['time'][0]['dataTime'])
    print (district['weatherElement'][0]['time'][0]['elementValue']['value'],district['weatherElement'][0]['time'][0]['elementValue']['measures'])
    print ('\n')
```
