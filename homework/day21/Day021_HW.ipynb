## 1. 取出今天所有的新聞

```python
from datetime import datetime
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
```

```python
# 打開瀏覽器


browser = webdriver.Chrome(executable_path='C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe')

browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
import time
for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 50000);")
```

```python
# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source
soup_today = BeautifulSoup(html_source, "html5lib")
```

```python
for d in soup_today.find(class_="part_list_2").find_all('h3'):  
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
FMT = '%Y/%m/%d %H:%M'
dt_start=soup_today.find(class_="part_list_2").find('h3').find(class_="date").text
dt_start=datetime.strptime(dt_start,FMT)
dt_start_stamp =dt_start.timestamp()
print(dt_start_stamp)
```

```python
for d in soup_today.find(class_="part_list_2").find_all('h3'):
    dt_next= d.find(class_="date").text
    dt_next=datetime.strptime(dt_next,FMT)
    dt_next_stamp =dt_next.timestamp()
    t_delta= dt_start_stamp-dt_next_stamp
    
    if t_delta<=7200:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
today_date=datetime.now()
```

```python
today_year=today_date.year
today_month=today_date.month
today_day=today_date.day
print('today is ',today_year,"/",today_month,"/",today_day)
```

```python
browser_3d = webdriver.Chrome(executable_path='C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe')

browser_3d.get("https://www.ettoday.net/news/news-list.htm")
selectSite = Select(browser_3d.find_element_by_id("selY"))
selectSite.select_by_value(str(today_year))

selectYear = Select(browser_3d.find_element_by_id("selM"))
selectYear.select_by_value(str(today_month))

selectYear = Select(browser_3d.find_element_by_id("selD"))
selectYear.select_by_value(str(today_day-3))

browser_3d.find_element_by_id('button').click()
```

```python
import time
for i in range(12):
    time.sleep(2)
    browser_3d.execute_script("window.scrollTo(0, 50000);")
```

```python
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```
