```python
import requests
```

```python
res = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
res.text
```

```python
import json
```

```python
forum1 = json.loads(res.text)
forum1
```

##這個 API 一次會回傳幾筆資料？

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

for article in forum1:
    print('標題:',article['title'])
    print('時間:',article['createdAt'])
    print('回文人數:',article['commentCount'])
    print('讚:',article['likeCount'])
    print('\n')
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
popular_count=0
popular_comment=0
for article in forum1:
    popular_count +=article['likeCount']
    popular_comment+=article['commentCount']
print('熱門文章')
print('平均按讚人數',popular_count/len(forum1))
print('平均留言人數',popular_comment/len(forum1))
```
